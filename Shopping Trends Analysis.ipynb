```python
import pandas as pd
import matplotlib.pyplot as plt
ds= pd.read_csv("C:/Users/shekh/Downloads/shopping_trends_analysis1.csv")

ds.groupby('Category')['Subscription Status'].count().plot(kind='bar', color='green', edgecolor='black')

plt.xlabel("Category", fontsize=12)
plt.ylabel("Number of Subscriptions", fontsize=12)
plt.title("Subscription Count by Category", fontsize=14)
plt.xticks(rotation=45)  # Adjust for better readability
plt.grid(axis='y', linestyle="--", alpha=0.5)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
ds= pd.read_csv("C:/Users/shekh/Downloads/shopping_trends_analysis1.csv")

ds.groupby('Item Purchased')['Discount Applied'].count().sort_values(ascending=False).plot(kind='bar', color='green', edgecolor='black')
plt.xlabel("Item Purchased", fontsize=12)
plt.ylabel("Number of Discounts Applied", fontsize=12)
plt.title("Total Discounts Given per Item", fontsize=14)
plt.xticks(rotation=45)  # Adjust if needed
plt.grid(axis='y', linestyle="--", alpha=0.5)
plt.xticks(rotation=45)  

plt.show()
```

```python
#Gender-wise Purchase Distribution:
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_csv("C:/Users/shekh/Downloads/shopping_trends_analysis1.csv")

ds['Gender'].value_counts().plot(kind='bar', color=['purple', 'black'], edgecolor='black')


plt.xlabel("Gender", fontsize=12)
plt.ylabel("Number of Purchases", fontsize=12)
plt.title("Distribution of Purchases by Gender", fontsize=14)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle="--", alpha=0.5)

plt.show()

```

```python
#sales trends changed over different years

sales_trends=ds.groupby('Year')['Purchase Amount (USD)'].sum().plot(kind='line', color='blue', marker='o')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Sales (USD)", fontsize=12)
plt.title("Sales Trends Over the Years", fontsize=14)
plt.grid(axis='y', linestyle="--", alpha=0.5)

plt.show()
```

```python
#Total Sales by Color

import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv("C:/Users/shekh/Downloads/shopping_trends_analysis1.csv")

color_sales = ds.groupby('Color')['Purchase Amount (USD)'].sum()

color_sales.plot(kind='bar', color='green', edgecolor='black',figsize=(10,5))

plt.xlabel("Purchased Color", fontsize=14)
plt.ylabel("Total Sales (USD)", fontsize=14)
plt.title("Total Sales by Color", fontsize=15)
plt.xticks(rotation=45)

plt.show()
```
